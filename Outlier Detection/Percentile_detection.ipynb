```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'../Datasets/student_dropout.csv')
df = df[['Age', 'Study_Hours_per_Day', 'Attendance_Rate', 'CGPA', 'Dropout']]
df = df.dropna()
df
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(1,2,1)
sns.distplot(df['Attendance_Rate'])

plt.subplot(1,2,2)
sns.boxplot(df['Attendance_Rate'])

plt.show
```

```python
max_limit = df['Attendance_Rate'].quantile(0.99)
min_limit = df['Attendance_Rate'].quantile(0.01)

print('Max limit: ', max_limit)
print('Min limit: ', min_limit)
```

```python
new_df = df[(df['Attendance_Rate'] < max_limit) & (df['Attendance_Rate'] > min_limit)]
```

```python
new_df
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(1,2,1)
sns.distplot(new_df['Attendance_Rate'])

plt.subplot(1,2,2)
sns.boxplot(new_df['Attendance_Rate'])

plt.show
```

#### Capping / Windsorization

```python
df['Attendance_Rate'] = np.where(df['Attendance_Rate'] >= max_limit,
        max_limit,
        np.where(df['Attendance_Rate'] <= min_limit,
        min_limit,
        df['Attendance_Rate']))
```
