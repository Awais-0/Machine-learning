```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'../Datasets/student_dropout.csv')
df = df[['Age', 'Study_Hours_per_Day', 'Attendance_Rate', 'CGPA', 'Dropout']]
df = df.dropna()
df
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(1,2,1)
sns.distplot(df['Attendance_Rate'])

plt.subplot(1,2,2)
sns.boxplot(df['Attendance_Rate'])

plt.show
```

```python
IQ25 = df['Attendance_Rate'].quantile(0.25)
IQ75 = df['Attendance_Rate'].quantile(0.75)

IQR = IQ75 - IQ25
```

```python
max_limit = IQ75 + 1.5*IQR
min_limit = IQ25 - 1.5*IQR
```

```python
print('Max_limit: ', max_limit)
print('Min limit: ', min_limit)
```

```python
new_df = df[df['Attendance_Rate'] > min_limit]
```

```python
new_df
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(1,2,1)
sns.distplot(new_df['Attendance_Rate'])

plt.subplot(1,2,2)
sns.boxplot(new_df['Attendance_Rate'])

plt.show
```
