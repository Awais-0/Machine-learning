```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r'../Datasets/placement_package.csv')
df
```

```python
df.iloc[112]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['cgpa'], df['package'], test_size=0.2, random_state=2)
```

```python
X_test
```

```python
y_test
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
print('Coefficient (m): ', lr.coef_)
print('Slope (b): ', lr.intercept_)
print('\n')
print('Following predictions are calculated using formula, y = mx + b \n')
for i in range(X_test.shape[0]):
    print(f'Actual (x,y) at {i}: ', X_test.iloc[i], ',',y_test.iloc[i])
    print(f'Predicted y at {i}: ',lr.predict([[X_test.iloc[i]]]))
```
