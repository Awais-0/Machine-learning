```python
class ChachuLR:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X_train, y_train):
        nume = 0
        deno = 0
        
        for i in range(X_train.shape[0]):
            nume = nume + ((y_train.iloc[i] - y_train.mean())*(X_train.iloc[i] - X_train.mean()))
            deno = deno + (X_train.iloc[i] - X_train.mean())*(X_train.iloc[i] - X_train.mean())
        self.m = nume / deno
        self.b = y_train.mean() - (self.m * X_train.mean())
        print('m: ', self.m)
        print('b: ', self.b)

    def predict(self, X_test):
        return self.m * X_test + self.b
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r'../Datasets/placement_package.csv')
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['cgpa'], df['package'], test_size=0.2, random_state=2)
```

```python
lr = ChachuLR()
```

```python
y_train[0]
```

```python
lr.fit(X_train, y_train)
```

```python
print(lr.predict(X_test))
```
