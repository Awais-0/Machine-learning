```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r'../../Datasets/titanic.csv')
df

df = df[['Survived', 'Age', 'Fare', 'Pclass']]
df
```

```python
df.isnull().mean()
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
df.isnull().mean()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']), df['Survived'], test_size=0.2, random_state=2)
```

```python
KNNimputer = KNNImputer(n_neighbors=3)
Simpleimputer = SimpleImputer()

X_train_KNN_trf = KNNimputer.fit_transform(X_train)
X_test_KNN_trf = KNNimputer.transform(X_test)

X_train_simple_trf = Simpleimputer.fit_transform(X_train)
X_test_simple_trf = Simpleimputer.transform(X_test)
```

```python
# pd.DataFrame(X_train_trf, columns=X_train.columns).isnull().mean()
pd.DataFrame(X_train_KNN_trf, columns=X_train.columns)
```

```python
lr1 = LogisticRegression()
lr2 = LogisticRegression()

lr1.fit(X_train_KNN_trf, y_train)
lr2.fit(X_train_simple_trf, y_train)

y_pred_KNN = lr1.predict(X_test_KNN_trf)
y_pred_simple = lr2.predict(X_test_simple_trf)

print('Accuracy score with KNN: ', accuracy_score(y_pred_KNN, y_test))
print('Accuracy score with simple: ', accuracy_score(y_pred_simple, y_test))
```

```python
for i in range(1, 11):
    imputer = KNNImputer(n_neighbors=i)
    X_train_KNN_trf = imputer.fit_transform(X_train)
    X_test_KNN_trf = imputer.transform(X_test)

    lr = LogisticRegression()
    lr.fit(X_train_KNN_trf, y_train)
    y_pred = lr.predict(X_test_KNN_trf)

    print(f'Accuracy score with {i} neighbors: ', accuracy_score(y_pred, y_test))
```
